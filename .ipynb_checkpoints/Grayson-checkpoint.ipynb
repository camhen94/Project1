```python
#import dependenciesimport pandas as pd
import pandas as pd
import numpy as np 
from pathlib import Path
import os
from scipy.stats import linregress
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
#import data and create data frame
#Get the file read into the notebook
nfl_file_path = Path("data_files/team_stats_2003_2023.csv")
nfl_data = pd.read_csv(nfl_file_path)
```

```python
#clean data
for row, team in enumerate(nfl_data["team"]):
    if team == "Oakland Raiders":
        nfl_data.loc[row, "team"] = "Las Vegas Raiders" 
    elif team == "St. Louis Rams":
        nfl_data.loc[row, "team"] = "Los Angeles Rams"
    elif team == "Washington Redskins":    
        nfl_data.loc[row, "team"] = "Washington Commanders"
    elif team == "Washington Football Team":
        nfl_data.loc[row, "team"] = "Washington Commanders"
    elif team == "San Diego Chargers":
        nfl_data.loc[row, "team"] = "Los Angeles Chargers"
```

```python
#clean data for my question
columns_to_remove = ['points_opp', 'g', 'turnovers', 'fumbles_lost', 'first_down', 'pass_cmp', 'pass_att', 'pass_yds', 'pass_td', 'pass_int', 'pass_net_yds_per_att', 'pass_fd', 'rush_att', 'rush_yds', 'rush_td', 'rush_yds_per_att' , 'rush_fd', 'penalties', 'penalties_yds', 'pen_fd' , 'score_pct', 'turnover_pct', 'exp_pts_tot', 'ties']
nfl_data = nfl_data.drop(columns_to_remove, axis=1)
display(nfl_data)
```

```python
# Define a function to create Linear Regression plots
def create_linear_regression_plot(x, y, x_label, y_label, title):
    # Perform linear regression
    slope, intercept, rvalue, pvalue, stderr = linregress(x, y)

    # Create scatter plot
    plt.scatter(x, y)

    # Create regression line
    regression_line = slope * x + intercept
    plt.plot(x, regression_line, color='red')

    # Incorporate the other graph properties
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)

    # Add trendline formula and R-value
    formula = f"y = {slope:.2f}x + {intercept:.2f}"
    text_box_props = dict(boxstyle='round', facecolor='white', edgecolor='gray', alpha=0.5)
    plt.text(x.min(), y.min(), formula, color='red', bbox=text_box_props, ha='left')

    # Save the figure
    #plt.savefig(save_file)
    
    # Print the R-value
    print(f"R-value is: {rvalue:.2f}")
    
    # Show plot
    plt.show()
```

```python
#win_loss_perc vs. total_yards
create_linear_regression_plot(nfl_data["total_yards"], nfl_data["win_loss_perc"], "Total Yards", "Win/Loss Percentage", "Total Yards vs. Win Percent")
create_linear_regression_plot(nfl_data["total_yards"], nfl_data["points_diff"], "Total Yards", "Point Differential", "Total Yards vs. Point Differential")
```

```python
#win_loss_perc vs. yds_per_play_offense
create_linear_regression_plot(nfl_data["yds_per_play_offense"], nfl_data["win_loss_perc"], "Yards Per Play", "Win/Loss Percentage", "Yards Per Play vs. Win Percent")
```

```python
#points_diff vs. yds_per_play_offense
create_linear_regression_plot(nfl_data["yds_per_play_offense"], nfl_data["points_diff"], "Yards Per Play", "Point Differential", "Yards Per Play vs. Point Differential")
```

```python
#points_diff vs. total_yards
create_linear_regression_plot(nfl_data["total_yards"], nfl_data["points_diff"], "Total Yards", "Point Differential", "Total Yards vs. Point Differential")
```

```python
#determine if high or low yards per play
average_yards_per_play = nfl_data['yds_per_play_offense'].mean()
nfl_data['YardsPerPlay_GreaterThanAvg'] = False
for index, row in nfl_data.iterrows():
    if row['yds_per_play_offense'] > average_yards_per_play:
        nfl_data.at[index, 'YardsPerPlay_GreaterThanAvg'] = True
display(nfl_data)        
```

```python
# Define a function to create Linear Regression plots
def create_linear_regression_plot(x, y, x_label, y_label, title, colors):
    # Perform linear regression
    slope, intercept, rvalue, pvalue, stderr = linregress(x, y)
    
    # Create scatter plot with color based on YardsPerPlay_GreaterThanAvg column
    plt.scatter(x, y, c=colors)
    
    # Create regression line
    regression_line = slope * x + intercept
    plt.plot(x, regression_line, color='red')
    
    # Incorporate the other graph properties
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    #plt.legend()
    
    # Add trendline formula and R-value
    formula = f"y = {slope:.2f}x + {intercept:.2f}"
    text_box_props = dict(boxstyle='round', facecolor='white', edgecolor='gray', alpha=0.5)
    plt.text(x.min(), y.min(), formula, color='red', bbox=text_box_props, ha='left')
    
    # Save the figure
    #plt.savefig(save_file)
    
    # Print the R-value
    print(f"R-value is: {rvalue:.2f}")
    
    # Show plot
    plt.show()
```

```python
colors = ['green' if val else 'red' for val in nfl_data['YardsPerPlay_GreaterThanAvg']]
create_linear_regression_plot(nfl_data["total_yards"], nfl_data["points_diff"], "Total Yards", "Point Differential", "Total Yards vs. Point Differential", colors)
create_linear_regression_plot(nfl_data["total_yards"], nfl_data["win_loss_perc"], "Total Yards", "Win/Loss Percentage", "Total Yards vs. Win/Loss Percentage", colors)
```

```python
# Step 1: Extract the relevant columns
yards_per_play_by_year = nfl_data[['year', 'yds_per_play_offense']]

# Step 2: Group the data by year and calculate the average yards per play
average_yards_per_play_by_year = yards_per_play_by_year.groupby('year').mean()

# Step 3: Create a line plot
plt.plot(average_yards_per_play_by_year.index, average_yards_per_play_by_year['yds_per_play_offense'])

# Step 4: Add trendline
slope, intercept, rvalue, pvalue, stderr = linregress(average_yards_per_play_by_year.index, average_yards_per_play_by_year['yds_per_play_offense'])
trendline = slope * average_yards_per_play_by_year.index + intercept
plt.plot(average_yards_per_play_by_year.index, trendline, color='red', label='Trendline')

plt.xlabel('Year')
plt.ylabel('Yards per Play')
plt.title('Yards per Play Over Time')
plt.show()
```

```python
# Step 2: Group the data by year and calculate the average yards per play
average_yards_per_play_by_year = yards_per_play_by_year.groupby('year').mean()
# Step 3: Create a boolean column indicating if yards per play is greater than the average
nfl_data['YardsPerPlay_GreaterThanAvg'] = nfl_data['yds_per_play_offense'] > nfl_data['year'].map(average_yards_per_play_by_year['yds_per_play_offense'])
# Step 4: assign color
colors = ['green' if val else 'red' for val in nfl_data['YardsPerPlay_GreaterThanAvg']]
```

```python
correlation_efficiency = nfl_data['win_loss_perc'].corr(nfl_data['yds_per_play_offense'])
correlation_total_yards = nfl_data['win_loss_perc'].corr(nfl_data['total_yards'])
abs_correlation_efficiency = abs(correlation_efficiency)
abs_correlation_total_yards = abs(correlation_total_yards)

if abs_correlation_efficiency > abs_correlation_total_yards:
    print("Efficiency (yds_per_play_offense) has a stronger correlation with win_loss_perc.")
else:
    print("Total yards has a stronger correlation with win_loss_perc.")
```

```python
partial_corr, _ = stats.pearsonr(nfl_data['win_loss_perc'] - nfl_data['total_yards'], nfl_data['yds_per_play_offense'])
if partial_corr > 0:
    print("When controlling for total yards, yds_per_play_offense has a positive correlation with win_loss_perc.")
elif partial_corr < 0:
    print("When controlling for total yards, yds_per_play_offense has a negative correlation with win_loss_perc.")
else:
    print("When controlling for total yards, there is no correlation between yds_per_play_offense and win_loss_perc.")
```

```python
partial_corr, _ = stats.pearsonr(nfl_data['win_loss_perc'] - nfl_data['total_yards'], nfl_data['plays_offense'])

if partial_corr > 0:
    print("When controlling for total yards, plays_offense has a positive correlation with win_loss_perc.")
elif partial_corr < 0:
    print("When controlling for total yards, plays_offense has a negative correlation with win_loss_perc.")
else:
    print("When controlling for total yards, there is no correlation between plays_offense and win_loss_perc.")
```

```python
correlation_efficiency = nfl_data['win_loss_perc'].corr(nfl_data['plays_offense'])
correlation_total_yards = nfl_data['win_loss_perc'].corr(nfl_data['total_yards'])
abs_correlation_efficiency = abs(correlation_efficiency)
abs_correlation_total_yards = abs(correlation_total_yards)

print("Correlation coefficient between win_loss_perc and plays_offense:", correlation_efficiency)
print("Correlation coefficient between win_loss_perc and total_yards:", correlation_total_yards)

if abs_correlation_efficiency > abs_correlation_total_yards:
    print("Efficiency (plays_offense) has a stronger correlation with win_loss_perc.")
else:
    print("Total yards has a stronger correlation with win_loss_perc.")
    
```
