```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
#Read in data from CSV as DataFrame
nfl_stats = Path("/Users/Tuna/Desktop/GitHub_Repo/Project1/team_stats_2003_2023.csv")
nfl_stats_df = pd.read_csv(nfl_stats)
```

```python
#Create new DataFrame for only relevant data
passing_game_df = nfl_stats_df[["year", "team", "wins", "losses", "pass_yds"]]
running_game_df = nfl_stats_df[["year", "team", "wins", "losses", "rush_yds"]]
merged_df = pd.merge(passing_game_df, running_game_df)
merged_df.head()

```

```python
#Clean DataFrame
merged_df = merged_df.rename(columns={"year": "Year", "team": "Teams", "wins": "Wins", "losses": "Losses", "pass_yds": "Passing Yards", "rush_yds": "Rushing Yards"})
merged_df.head()
```

```python
# Compute pearson correlation coefficient
pearson_corr = merged_df[["Rushing Yards", "Passing Yards", "Wins", "Losses"]].corr(method='pearson')

print("Pearson Correlation:")
print(pearson_corr)

```

```python
#Create scatter plot for Rushing Yards vs. Wins
plt.figure(figsize=(10, 6))
plt.scatter(merged_df["Rushing Yards"], merged_df["Wins"], alpha=0.5)
plt.title('Rushing Yards vs. Wins')
plt.xlabel('Rushing Yards')
plt.ylabel('Wins')
plt.grid(True)
plt.show()

#Create scatter plat for Passing Yards vs. Wins
plt.figure(figsize=(10, 6))
plt.scatter(merged_df["Passing Yards"], merged_df["Wins"], alpha=0.5)
plt.title('Passing Yards vs. Wins')
plt.xlabel('Passing Yards')
plt.ylabel('Wins')
plt.grid(True)
plt.show()
```
